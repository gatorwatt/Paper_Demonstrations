# C - Assigning Transforms and Infill

Assigning root categories is conducted in assigncat parameter, assigning infill in assigninfill, and parameters in assignparam - e.g. for a train set df_train with column headers 'col1' and 'col2' we could assign string parsing (splt) and a string parse family tree (or19) with infill types zero infill and ML infill.

Since 'splt' transform accepts parameters, we'll also demonstrate passing parameter of excluding space and special characters, such as to promote single word overlap detections.

Note any columns we don't explicitly assign will defer to automation or we could turn off automated defaults for pass-through of other columns by passing automunge parameter powertransform = 'excl'.

```python
#Note our sample test set doesn't include columns titled 'col1' and 'col2'
#So in order to match the conventions of appendix let's just create such columns
#Selected randomly from the categoric entries in this set

df_train['col1'] = df_train['LotConfig'].copy()
df_train['col2'] = df_train['Neighborhood']

df_test['col1'] = df_test['LotConfig'].copy()
df_test['col2'] = df_test['Neighborhood']

#As an aside, the default minsplit parameter for string parsing functions is 5
#Meaning overlaps below this threshold are not registered
#This value can be reconfigured in the assignparam
#See the other demonstration notebook 
#"String_Theory_demonstrations_from_paper
```

```python
train, train_ID, labels, \
val, val_ID, val_labels, \
test, test_ID, test_labels, \
postprocess_dict \
= am.automunge(df_train, 
  assigncat = {'splt':['col1'], 'or19':['col2']},
  assigninfill = {'zeroinfill':['col1'], 'MLinfill':['col2']},
  assignparam = {'splt' : {'col1' :
                     {'space_and_punctuation' : False}}})
```

# D - Overwriting Sets of Transforms

Here we'll demonstrate overwriting sets of transformations with the transformdict parameter passed to automunge(.).

Let's use the example from paper of 'or19' application of 'nmr7' to extract numerical portions which has result of returning the zip codes in the address examples. We noted in paper that there might be benefit to encoding the output as categorical instead of a numeric (z-score) normalization, so let's demonstrate overwriting the pre-defined transform to apply a categoric encoding downstream of the 'nmr7'.

Referring to Figure 7, we see the relevant family tree is for category 'nmc8' which has a children primitive entry of 'nmbr'. These family trees are also available for inspection in the READ ME.

So if we want to instead encode the numeric extract as categorical, we could overwrite 'nmc8' in a custom passed transformdict.

Note that since 'nmc8' is called as part of an offspring generation, the 'nmc8' upstream primitives (parents / siblings / auntsuncles / cousins) aren't inspected in context of 'or19', so we only have to worry about the downstream primitives (children / niecesnephews / coworkers / friends).

Let's demonstrate the overwrite, we'll instead use 'ord3' for ordinal encoding by frequency.

```python
transformdict =  {'nmc8' : {'parents'       : ['nmc8'], 
                            'siblings'      : [], 
                            'auntsuncles'   : [], 
                            'cousins'       : ['NArw'], 
                            'children'      : ['ord3'], 
                            'niecesnephews' : [], 
                            'coworkers'     : [], 
                            'friends'       : []}}
```

And since we are using existing categories from the library, we don't need to repopulate a corresponding processdict.

Then when we call automunge(.) we can pass this transformdict to overwrite 'nmc8':

```python
train, train_ID, labels, \
val, val_ID, val_labels, \
test, test_ID, test_labels, \
postprocess_dict \
= am.automunge(df_train, 
  assigncat = {'or19':['col2']}, 
  transformdict = transformdict)
```

The returned columns log the steps of transformations via suffix appenders, so this replaces the original returned column 'col2_UPCS_nmc7_nmbr' with 'col2_UPCS_nmc7_ord3'.

```python
#to validate, here is confirmation
if 'col2_UPCS_nmc7_nmbr' not in list(train):
  print("confirmed 'col2_UPCS_nmc7_nmbr' not in finalcolumns_train")
  print()
  
if 'col2_UPCS_nmc7_ord3' in list(train):
  print("confirmed 'col2_UPCS_nmc7_ord3' in finalcolumns_train")
  print()
```

# E - Feature Importance

```python
#Here's we'll demosnrtate an automunge(.) call with feature importance

#since this method needs a label column let's reload sets (we dropped column earlier in notebook)
df_train = pd.read_csv('housing_train.csv')
df_test = pd.read_csv('housing_test.csv')

#by inspection these columns are labels and index
labels_column = 'SalePrice'
ID_columnlist = 'Id'

train, train_ID, labels, \
val, val_ID, val_labels, \
test, test_ID, test_labels, \
postprocess_dict \
= am.automunge(df_train, 
               labels_column=labels_column, 
               trainID_column=ID_columnlist, 
               featureselection=True)

#if you scroll through the printouts you'll find results about midway
```

```python
#alternately we could just run a report such that results will be at conclusion of printouts

train, train_ID, labels, \
val, val_ID, val_labels, \
test, test_ID, test_labels, \
postprocess_dict \
= am.automunge(df_train, 
               labels_column=labels_column, 
               trainID_column=ID_columnlist, 
               featureselection='report')
```

```python
#note the results are also available in returned postprocess_dict

#to inspect values returned in featureimportance object one could run
for keys,values in postprocess_dict['FS_sorted']['column_key'].items():
    print(keys)
    print(values)
    print()
```

```python

#here are the relative metric reults for instance
postprocess_dict['FS_sorted']['metric2_column_key']

#the results have first tier of the source featrue column header, 
#second tier of returned column header
#and corresponding value of the metric2 results
#such as for comparison between columns dervied from the same feature

#Note that in a full automunge(.) call these are stored under key 'FS_sorted' eg
#postprocess_dict['FS_sorted']['metric2_column_key']
```

```python
#we noted in the appendix that the relative metrics can be useful for evaluating
#influence of different segments of a feature's distribution
#here for example we apply a 'text' one-hot encoding to the feature GarageCars
#note that these relative metrics are as much as measure of the model as they are of the feature

#here we turn up the n_estimators which should improve model performance

train, train_ID, labels, \
val, val_ID, val_labels, \
test, test_ID, test_labels, \
postprocess_dict \
= am.automunge(df_train, \
               labels_column = labels_column, \
               trainID_column = ID_columnlist, \
               shuffletrain = False, \
               MLinfill=False,
               printstatus = False, \
               pandasoutput = True, \
               powertransform = False, \
               assigncat = {'text':['GarageCars']}, \
               featureselection = True, \
               ML_cmnd = {'autoML_type':'randomforest', \
                          'MLinfill_cmnd':{'RandomForestClassifier':{'n_estimators':1000}, \
                                           'RandomForestRegressor':{'n_estimators':1000}}}
               )

#Here we see that two car garages
#are more influential to predictive accuracy
#than other variations.
postprocess_dict['FS_sorted']['metric2_column_key']['GarageCars']

#Here we see that two car garages are more influential to this model
```

```python
#we also noted that a similar approach could be applied to evaluate 
#influence of different segments of a numeric variables distribution
#this takes a 'tlbn' encoding further detauiled in the READ ME

train, train_ID, labels, \
val, val_ID, val_labels, \
test, test_ID, test_labels, \
postprocess_dict \
= am.automunge(df_train, \
               labels_column = labels_column, \
               trainID_column = ID_columnlist, \
               shuffletrain = False, \
               printstatus = False, \
               pandasoutput = True, \
               powertransform = False, \
               MLinfill=False,
               assigncat = {'tlbn':['LotArea']}, \
               assignparam = {'tlbn' : {'LotArea' : {'bincount' : 5}}}, \
               featureselection = True, \
               ML_cmnd = {'autoML_type':'randomforest', \
                          'MLinfill_cmnd':{'RandomForestClassifier':{'n_estimators':1000}, \
                                           'RandomForestRegressor':{'n_estimators':1000}}})
               
postprocess_dict['FS_sorted']['metric2_column_key']['LotArea']

#this results demosnrtates that for equal population segments of the distribution
#the last bin (bin 4) with largest values is the most influential
#and the first bin (bin 0) with smallest values is the second most influential
#to this particular model
#note that the number of bins is configurable 
```
